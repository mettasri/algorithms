### 1. Two Sum

```python
def twoSum(nums, target):
    numMap = {}
    for i, num in enumerate(nums):
        first = target - num
        if first in numMap:
            return [numMap[first], i]
        numMap[num] = i
```

```python
print (twoSum([2, 7], 9))
print (twoSum([2, 7, 11, 15], 17))
```

### 7. Reverse Integer

```python
def reverse(x):
    # Get the sign of integer
    sign = -1 if x < 0 else 1
    x = abs(x)
    
    # Reverse the integer
    res = 0
    while (x):
        res = res*10 + x%10
        x = int(x/10)

    res = sign*res
    # check for the integer overflow
    if (res < -pow(2, 31) or res > (pow(2, 31) - 1)):
        return 0
    return res
```

```python
def isPalindrome(n):
    # negative number or number ends with zero
    if n < 0 or (n%10 == 0 and n != 0):
        return False
    
    # split the number into two equal halfs
    # eg: n = 12321; n = 12, rev_n = 123
    # eg: n = 1221;  n = 12, rev_n = 12
    rev_n = 0
    while (n > rev_n):
        rev_n = rev_n*10 + n%10
        n = int(n/10)
    #print (n, rev_n)
    return n == rev_n or n == int(rev_n/10)
```

### 13. Roman to Integer

```python
def romanToInt(s):
    rtoi = {'I': 1, 'V': 5, 'X': 10, 'L': 50, 'C': 100, 'D': 500, 'M': 1000}
    res, last = 0, 0
    
    for ch in s:
        cur = int(rtoi[ch])
        res += cur
        if last < cur:
            res -= last*2
        last = cur
    return res
```

```python
import sys
def romanToInt(s):
    rtoi = {'I': 1, 'V': 5, 'X': 10, 'L': 50, 'C': 100, 'D': 500, 'M': 1000}
    if s == "":
        return 0
    if len(s) == 1:
        return rtoi[s[0]]
    
    num = rtoi[s[-1]]
    for i in range(len(s))[:-1]:
        sign = -1 if rtoi[s[i]] < rtoi[s[i + 1]] else 1
        num += sign * rtoi[s[i]]
    return num
```

### 14. Longest Common Prefix

```python
# Approach 1: Horizontal Scanning
#    get the first string
#    for each remaining strings
#       find the common prefix
# Time complexity  : O(S); S sum of all characters in all strings
# Space complexity : O(1)

import re
def longestCommonPrefix(strs):
    if strs == "" or len(strs) == 0:
        return ""
    
    prefix = strs[0]
    for i in range(1, len(strs)):
        while re.match(prefix, strs[i]) == None:
            prefix = prefix[:-1]
            if prefix == "":
                return ""
    return prefix
```

```python
# Approach 2: Vertical Scanning
#    for each char from first string
#        compare with the char at the same position in other strings
# Time complexity
#        worst case: O(m * n); m is length, n is num of strings
#        best case : O(n * minLen); minLen is small string of all strings
# Space complexity : O(1)

def longestCommonPrefix(strs):
    if strs == "" or len(strs) == 0:
        return ""
    
    for col in range(len(strs[0])):
        for row in range(1, len(strs)):
            if col < len(strs[row]) and strs[0][col] != strs[row][col]:
                return strs[0][:col]
    return strs[0]
```

```python
# Approach 3: Devide & Conquer
#    get common prefix from the two strings by deviding into groups
#    conquer the results finally
# Time complexity
#        worst case: O(m * n); m is length, n is num of strings
#        best case : O(minLen * n)
# Space complexity : O(m log n); log n recursive call, each call needs m space to store the result

def longestCommonPrefix(strs):
    if strs == "" or len(strs) == 0:
        return ""

    return longestCommonPrefixUtil(strs, 0, len(strs) - 1)
    
def longestCommonPrefixUtil(strs, l, h):
    if l == h:
        return strs[l]
    
    m = int((l + h)/2)
    left_prefix = longestCommonPrefixUtil(strs, l, m)
    right_prefix = longestCommonPrefixUtil(strs, m + 1, h)
    return commonPrefix(left_prefix, right_prefix)

def commonPrefix(str1, str2):
    min_len = min(len(str1), len(str2))
    for i in range(min_len):
        if (str1[i] != str2[i]):
            return str1[:i]
    return str1[:min_len]
```

```python
# Approach 4: Binary Search
#    get the smallest string length out of all the strings
#    search that present in all strings in binary search
# Time complexity  : O(S * log n); S length of all strings, n num of strings
# Space complexity : O(1)

import sys
def longestCommonPrefix(strs):
    if strs == "" or len(strs) == 0:
        return ""
    
    min_len = sys.maxsize
    for s in strs:
        min_len = min(min_len, len(s))

    l = 1
    h = min_len
    while(l <= h):
        m = int((l + h)/2)
        if isCommonPrefix(strs, m):
            l = m + 1
        else:
            h = m - 1
    length = int((l + h)/2)
    return strs[0][:length]

def isCommonPrefix(strs, length):
    str1 = strs[0][:length]
    for i in range(1, len(strs)):
        if re.match(str1, strs[i][:length]) == None:
            return False
    return True
```

```python
# Approach 5: Using Trie data structure
# Time complexity
#         preprocessing O(S); S is number of all characters in array
#         LCP query : O(m); m is query length
# Space complexity : O(S)

class TrieNode:
    def __init__(self):
        self.children = [None]*26
        self.isEnd = False

class Trie:
    def __init__(self):
        self.root = self.getNode()
    
    def getNode(self):
        return TrieNode()
        
    def insert(self, string):
        if string == "":
            return
        
        crawl = self.root
        for ch in string:
            index = ord(ch) - ord('a')
            if not crawl.children[index]:
                crawl.children[index] = self.getNode()
            crawl = crawl.children[index]
        crawl.isEnd = True

    def longestCommonPrefix(self):
        return self.commonPrefix(self.root, "")
    
    # each node should have only one child
    # isEnd reached
    def commonPrefix(self, crawl, prefix):
        if crawl.isEnd:
            return prefix
        
        # only one child
        child_count = 0
        for i in range(26):
            if crawl.children[i]:
                child_count += 1
            if child_count > 1:
                return prefix
        
        for i in range(26):
            if crawl.children[i]:
                ch = chr(i + ord('a'))
                return self.commonPrefix(crawl.children[i], prefix + ch)
        return ""

def longestCommonPrefix(strs):
    if strs == "" or len(strs) == 0:
        return ""
    
    t = Trie()
    for string in strs:
        #print("inserting:", string)
        if string == "":
            return ""
        t.insert(string)
    
    return t.longestCommonPrefix()
```

### 20. Valid parentheses

```python
def isValidParentheses(s):
    right_parens = {'}' : '{', ')' : '(', ']' : '['}
    stack = []
    for paren in s:
        if paren not in right_parens.keys():
            stack.append(paren)
        else: #if paren in right_parens.keys():
            if not stack or right_parens[paren] != stack.pop():
                return False
    return not stack
```

```python
print(isValidParentheses("()")) # True
print(isValidParentheses("(")) # False
print(isValidParentheses("()[]{}")) # True
print(isValidParentheses("(]")) # False
print(isValidParentheses("{[]}")) # True
print(isValidParentheses("){")) # False
print(isValidParentheses("")) # True
```

### 26. Remove Duplicates from Sorted Array

```python
def removeDuplicates(nums):
    if len(nums) < 2: return nums
    
    w = 1
    for r in range(1, len(nums)):
        if nums[r] != nums[w - 1]:
            nums[w] = nums[r]
            w += 1

    return nums[ : w]
```

### 28. Implement strStr()

```python
def strStr(haystack, needle):
    if needle == "": return 0
    if len(haystack) < len(needle): return -1
    
    for i in range(len(haystack) - len(needle) + 1):
        if (haystack[i : i + len(needle)] == needle):
            return i
    return -1
```

```python
print(strStr("hello", "ll"))
print(strStr("aaaaaa", "bba"))
print(strStr("aaa", "aaaaa"))
print(strStr("mississippi", "issipi"))
print(strStr("a", "a"))
```

### 35. Search Insert position

```python
def searchInsert(nums, target):
    low = 0
    high = len(nums) - 1
    while (low <= high):
        mid = (low + high)//2
        if (target == nums[mid]):
            return mid
        
        if (target < nums[mid]):
            high = mid - 1
        elif (target > nums[mid]):
            low = mid + 1
    #print (low, high)
    return (low + high)//2 + 1
```

### 53. Maximum Subarray

```python
import sys
def maxSubArray(nums):
    maxSum = -sys.maxsize - 1
    curSum = 0
    for num in nums:
        curSum = max(curSum + num, num)
        maxSum = max(maxSum, curSum)
    return maxSum
```

### 58. Length of Last Word

```python
def lengthOfLastWord(s):
    wordEnd = -1
    for i in range(len(s))[::-1]:
        if s[i] == ' ' and wordEnd != -1:
            return wordEnd - i
        elif s[i] != ' ' and wordEnd == -1:
            wordEnd = i
    return wordEnd + 1
```

```python
def mySqrt(x):
    lo, hi = 1, x/2 + 1
    while lo <= hi:
        mi = (lo + hi)//2
        exp = mi * mi
        if x > exp:
            lo = mi + 1
        elif x < exp:
            hi = mi - 1
        else:
            return mi
    return hi
```

### 88. Merge Sorted Array

```python
def merge(nums1, m, nums2, n):
    while (m > 0 and n > 0):
        if (nums1[m - 1] <= nums2[n - 1]):
            nums1[m + n - 1] = nums2[n - 1]
            n -= 1
        else:
            nums1[m + n - 1] = nums1[m - 1]
            m -= 1
    if (n > 0):
        nums1[:n] = nums2[:n]
```

### 125 Valid Palindrome

```python
def isPalindrome(s):
    l, h = 0, len(s) - 1
    while l < h:
        #print(s[l], s[h])
        if not s[l].isalnum():
            l += 1
            continue
        if not s[h].isalnum():
            h -= 1
            continue
        if s[l].lower() != s[h].lower():
            return False
        l +=1
        h -= 1
    return True
```

```python
print(isPalindrome("A man, a plan, a canal: Panama")) # True
print(isPalindrome("mala ya, la, m")) # True
print(isPalindrome("race a car")) # False
print(isPalindrome("0P")) # False
print(isPalindrome("p,,, ")) # True
print(isPalindrome(",,, ")) # True
```

### 155. Min Stack

```python
class MinStack:

    def __init__(self):
        """
        initialize your data structure here.
        """
        self.data = []
        self.mindata = []

    def push(self, x):
        """
        :type x: int
        :rtype: void
        """
        self.data.append(x)
        
        if self.mindata == []:
            self.mindata.append(x)
        else:
            self.mindata.append(min(x, self.mindata[-1]))

    def pop(self):
        """
        :rtype: void
        """
        self.data.pop()
        self.mindata.pop()
        return
    
    def top(self):
        """
        :rtype: int
        """
        return self.data[-1]

    def getMin(self):
        """
        :rtype: int
        """
        if self.mindata == []:
            return float("inf")
        return self.mindata[-1]
```

### 167. Two Sum II - Input array is sorted

```python
def twoSum(numbers, target):
    l, h = 1, len(numbers)
    while l < h:
        if (numbers[l - 1] + numbers[h - 1]) < target:
            l += 1
        elif (numbers[l - 1] + numbers[h - 1]) > target:
            h -= 1
        else:
            return l, h
```

```python
# Divide and Conquer
# Time complexity O(n lg n)
# Space complexity O(lgn)
def majorityElementRec(nums, lo, hi):#, ind):
    if lo == hi:
        #print(" "*ind, lo, hi, nums[lo])
        return nums[lo]

    mi = (hi - lo)//2 + lo
    lh = majorityElementRec(nums, lo, mi)#, ind + 6)
    rh = majorityElementRec(nums, mi + 1, hi)#, ind + 6)
    if lh == rh:
        #print(" "*ind, lo, hi, lh, rh)
        return lh
    lc = sum(1 for i in range(lo, hi + 1) if nums[i] == lh)
    rc = sum(1 for i in range(lo, hi + 1) if nums[i] == rh)
    lrm = lh if lc > rc else rh
    #print(" "*ind, lo, hi, lrm)
    return lrm
    
def majorityElement(nums):
    return majorityElementRec(nums, 0, len(nums) - 1)#, 1)
```

```python
# Boyer-Moore Voting Algorithm
# Time complexity O(n)
# Space complexity O(1)
def majorityElement(nums):
    candidate = None
    count = 0
    for i in range(len(nums)):
        if count == 0:
            candidate = i
        count += (1 if nums[i] == nums[candidate] else -1)
    return nums[candidate]
```

### 172. Factorial Trailing Zeroes
All trailing zeros are come from even_num x 5, we have more even_num than 5, so only count factor 5.

4! = 1x2x3x4 = 24, we havent encountered any 5 yet, so we dont have any trailing zero.

5! = 1x2x3x4x5 = 120, we have one trailing zero. either 25, or 45 can contribute to that zero.

9! = 362880, we only encountered 5 once, so 1 trailing zero as expected.

10! = 3628800, 2 trailing zeros, since we have two numbers that have factor 5, one is 5 and the other is 10 (25)

What about 100! then?

100/5 = 20, we have 20 numbers have factor 5: 5, 10, 15, 20, 25, , 95, 100.

Is the number of trailing zero 20? No, its 24, why?

Within that 20 numbers, we have 4 of them: 25 (55), 50 (2x5x5), 75 (3x5x5), 100 (4x5x5) that have an extra factor of 5.

So, for a given number n, we are looking how many numbers <=n have factor 5, 55, 5x5x5, 

Summing those numbers up we got the answer.

e.g. 1000! has 249 trailing zeros:

1000/5 = 200

1000/25 = 40

1000/125 = 8

1000/625 = 1

200 + 40 + 8 + 1 = 249

alternatively, we can do the following

1000/5 = 200

200/5 = 40

40/5 = 8

8/5 = 1

1/5 = 0

200 + 40 + 8 + 1 + 0 = 249



```python
def trailingZeroes(n):
    count = 0
    while n > 0:
        n = n//5
        count += n
    return count
```

### 189. Rotate Array

```python
# Brute Force
# Time complexity O(n*k)
# Space complexity O(1)
def rotate(nums, k):
    while(k > 0):
        last = nums[-1]
        for i in range(len(nums)-1, 0, -1):
            nums[i] = nums[i-1]
        nums[0] = last
        k -= 1
```

```python
# Using Reverse
# Time complexity O(n), three passes
# Space complexity O(1)
def rotate(nums, k):
    if len(nums) < 2: return
    k = k % len(nums)
    if k == 0: return
    
    reverse(nums, 0, len(nums)) # [7,6,5,4,3,2,1]
    reverse(nums, 0, k)         # [5,6,7,4,3,2,1]
    reverse(nums, k, len(nums)) # [5,6,7,1,2,3,4]
    
def reverse(nums, lo, hi):
    while lo < hi:
        nums[lo], nums[hi - 1] = nums[hi - 1], nums[lo]
        lo += 1
        hi -= 1
```

### 190. Reverse Bits

```python
def reverseBits(n):
    ans = 0
    for i in range(32):
        set_mask = (n >> i) & 1
        if set_mask != 0:
            ans |= set_mask << (31 - i)
    return ans
```

### 198. House Robber

```python
# dynamic programming
def rob(nums):
    prevMax = 0
    curMax = 0
    for num in nums:
        newMax = max(prevMax + num, curMax)
        prevMax = curMax
        curMax = newMax
    return curMax
```

```python
print(rob([100,79,100,1000])) # 1100
print(rob([1,2,3,1]))   # 4
print(rob([2,7,9,3,1])) # 12
print(rob([2,1,1,2]))   # 4
```

### 205. Isomorphic Strings

```python
def isIsomorphic(s, t):
    if len(s) != len(t):
        return False
    
    isoMap = {}
    for i in range(len(s)):
        if s[i] in isoMap:
            if isoMap[s[i]] != t[i]:
                return False
        elif t[i] in isoMap.values():
            return False
        else:
            isoMap[s[i]] = t[i]
    return True
```

### 219. Contains Duplicate II

```python
def containsNearbyDuplicate(nums, k):
    indexMap = {}
    for i, num in enumerate(nums):
        if num in indexMap and (i - indexMap[num]) <= k:
            return True
        indexMap[num] = i
    return False
```

### 225. Implement Stack using Queues

```python
# 1) Two Queues, push - O(1), pop O(n)
# 2) Two Queues, push - O(n), pop O(1)

# 3) One Queue, push - O(n), pop O(1)
class MyStack:

    def __init__(self):
        """
        Initialize your data structure here.
        """
        self.queue = []

    def push(self, x):
        """
        Push element x onto stack.
        :type x: int
        :rtype: void
        """
        self.queue.append(x)
        size = len(self.queue)
        while size > 1:
            self.queue.append(self.queue.pop(0))
            size -= 1

    def pop(self):
        """
        Removes the element on top of the stack and returns that element.
        :rtype: int
        """
        if len(self.queue) > 0:
            return self.queue.pop(0)

    def top(self):
        """
        Get the top element.
        :rtype: int
        """
        if len(self.queue) > 0:
            return self.queue[0]

    def empty(self):
        """
        Returns whether the stack is empty.
        :rtype: bool
        """
        return self.queue == []


# Your MyStack object will be instantiated and called as such:
# obj = MyStack()
# obj.push(x)
# param_2 = obj.pop()
# param_3 = obj.top()
# param_4 = obj.empty()
```

### 232. Implement Queue using Stacks

```python
# (Two Stacks) Push - O(n) per operation, Pop - O(1) per operation
class MyQueue:

    def __init__(self):
        """
        Initialize your data structure here.
        """
        self.stack = []

    def push(self, x):
        """
        Push element x to the back of queue.
        :type x: int
        :rtype: void
        """
        auxStack = []
        while len(self.stack) > 0:
            auxStack.append(self.stack.pop(-1))

        self.stack.append(x)
        while len(auxStack) > 0:
            self.stack.append(auxStack.pop(-1))

    def pop(self):
        """
        Removes the element from in front of queue and returns that element.
        :rtype: int
        """
        if len(self.stack) == 0:
            return None
        
        return self.stack.pop(-1)
    
    def peek(self):
        """
        Get the front element.
        :rtype: int
        """
        if len(self.stack) == 0:
            return None
        
        return self.stack[-1]

    def empty(self):
        """
        Returns whether the queue is empty.
        :rtype: bool
        """
        return len(self.stack) == 0


# Your MyQueue object will be instantiated and called as such:
# obj = MyQueue()
# obj.push(x)
# param_2 = obj.pop()
# param_3 = obj.peek()
# param_4 = obj.empty()
```

### 243. Shortest Word Distance

```python
def shortestDistance(words, word1, word2):
    found = None
    shortest = len(words)
    for i in range(len(words)):
        if word1 == words[i] or word2 == words[i]:
            if found != None and words[i] != words[found]:
                shortest = min(abs(i - found), shortest)
            found = i
    return shortest
```

```python
import sys
def minCost(costs):
    houseCount = len(costs)
    if houseCount == 0:
        return 0
    for i in range(1, houseCount):
        for j in range(3):
            j1 = (j + 1) % 3
            j2 = (j + 2) % 3
            costs[i][j] = min(costs[i][j] + costs[i - 1][j1], costs[i][j] + costs[i - 1][j2])
    return min(costs[houseCount - 1])
```

### 266. Palindrome Permutation

```python
def canPermutePalindrome(s):
    charMap = {}
    for ch in s:
        charMap[ch] = charMap.get(ch, 0) + 1
    
    oddCounts = 0
    for _, value in charMap.items():
        if value % 2:
            oddCounts += 1
        if oddCounts > 1:
            return False
    return True
```

```python
def canPermutePalindrome(s):
    count = {}
    oddCount = 0
    for ch in s:
        count[ch] = count.get(ch, 0) + 1
        if count[ch] % 2 == 0:
            oddCount -= 1
        else:
            oddCount += 1
    
    #print(oddCount)
    return oddCount <= 1
```

```python
from collections import Counter
def canPermutePalindrome(s):
    count = Counter(s)
    
    oddCounts = 0
    for _, value in count.items():
        if value % 2:
            oddCounts += 1
        if oddCounts > 1:
            return False
    return True
```

```python
print(canPermutePalindrome("code"))
print(canPermutePalindrome("aab"))
print(canPermutePalindrome("carerac"))
```

### 268. Missing Number

```python
def missingNumber(nums):
    missing = 0
    for num in nums:
        missing ^= num
    for num in range(len(nums) + 1):
        missing ^= num
    return missing
```

```python
def missingNumber(nums):
    missing = len(nums)
    for i, n in enumerate(nums):
        missing ^= i ^ n
    return missing
```

```python
print(missingNumber([3,0,1]))
print(missingNumber([9,6,4,2,3,5,7,0,1]))
print(missingNumber([9,6,4,2,8,5,7,0,1]))
```

### 276. Paint Fence

```python
def numWays(n, k):
    """
    :type n: int
    :type k: int
    :rtype: int
    """
    if n == 0 or k == 0: return 0
    if n > k and k < 2: return 0
    n, k = max(n, k), min(n, k)
    ways = 1
    for i in range(n, n-k, -1):
        ways *= i
    return ways
```

```python
print (numWays(0, 0)) # 0
print (numWays(1, 0)) # 0
print (numWays(0, 1)) # 0
print (numWays(1, 1)) # 1
print (numWays(3, 2)) # 6
print (numWays(3, 1)) # 0
print (numWays(1, 2)) # 2
print (numWays(1, 6)) # 6
print (numWays(2, 1)) # 1
print (numWays(6, 1)) # 0
```

### ncr

```python
def ncr(n, r):
    if n < r:
        return "n should be smaller than r"
    return int(fact(n) / (fact(r) * fact(n - r)))

def fact(n):
    if n < 1:
        return 1
    return n * fact(n - 1)
```

```python
print(ncr(1, 1))
print(ncr(2, 1))
print(ncr(5, 3))
print(ncr(5, 2))
print(ncr(5, 6))
print(ncr(0, 0))
```
